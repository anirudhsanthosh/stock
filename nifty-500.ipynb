```python
import yfinance as yf
import pandas as pd 
import datetime
from dateutil.relativedelta import relativedelta
```

```python
index = pd.read_csv("nifty-500.csv", index_col=False)
```

```python
nify_500 = index["Symbol"][101:]
```

```python
returns = {}

for symbol in nify_500:

    ticker = symbol+".NS"

    print(symbol)

    price = yf.download(ticker, start="2022-11-01", end="2022-11-04")

    price.drop(["Adj Close", "Volume"], axis=1, inplace=True)

    month_open = price.iloc[0]["Open"]
    month_close = price.iloc[-1]["Close"]


    returns[symbol] = (month_close / month_open) - 1
```

```python
returns = {k: v for k, v in sorted(returns.items(), key=lambda item: item[1])}
```

```python
for k, v in returns.items():

    print(k, v * 100)
```
